My dataset is from Kaggle, find it here: https://www.kaggle.com/datasets/josephvm/futurama-seasons-16-transcripts

```python
#import libraries

import pandas as pd
```

```python
df = pd.read_csv('futurama_spoken_text.csv')
```

Huzzah! The data is in and now we can analyze.

Alrighty, we only need to know what Bender's lines are. Let's filter down and see what we've got.

```python
df.value_counts('Character')
```

```python
bender = df[df['Character'].str.contains("Bender")]
```

```python
bender
```

```python
bender.Character.value_counts()
```

Okay we have a dataset with only Bender's lines, but now we need to clean this up more. All Benders should just be called Bender.(not Bonder)

```python
errors='ignore'
bender.loc[bender['Character'].str.contains('Bender', na=False), 'Character'] = 'Bender'
```

```python
bender.Character.value_counts()
```

```python
bender.Line.value_counts()
```

Much better!
